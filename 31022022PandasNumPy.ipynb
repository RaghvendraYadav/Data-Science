```python
def profit_flag(a):
    if a>=0:
        return 'positive'
    else:
        return 'negative'
```

```python
profit_flag(-12)
```

```python
profit_flag(12)
```

```python
df_sales['flag_profit']=df_sales['profit'].apply(profit_flag)# Create Derived Column using apply
```

```python
df_sales.head()
```

```python
df_sales['Customer_Name_Length']=df_sales['customer_name'].apply(len)
```

```python
df_sales.head()
```

```python
def Quantity_type(a):
    if a<10:
        return 'low'
    elif a>10 and a<20:
        return 'Medium'
    else:
        return 'High'
```

```python
df_sales['Flag_Quantity']=df_sales['quantity'].apply(Quantity_type)
```

```python
df_sales['Flag_Quantity'].value_counts()
```

```python
df_sales['square_quantity']=df_sales['quantity'].apply(lambda a:a**2)
```
