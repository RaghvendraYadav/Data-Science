```python
# Reading the Dataset in Pandas
import pandas as pd
df=pd.read_csv('FitBit data.csv')
df
```

```python
# Reading the Dataset in MongoDB
import pymongo
Client=pymongo.MongoClient('mongodb://localhost:27017')
db=Client['FitBitData']
data=df.to_dict(orient='records')
collection=db['FitBitCollection']
collection.insert_many(data)
```

```python
# Reading Data in mysql
import mysql.connector as conn
from sqlalchemy import engine
```

```python
# Read Data From SQL
my_db=conn.connect(host='localhost',user='root',password='0000')
cursor=my_db.cursor()
cursor.execute('create database FitBitData')
```

```python
#Engine Creation
engine=engine.create_engine('mysql+pymysql://root:0000@localhost/FitBitData')
```

```python
df.to_sql('FitBitData',con=engine)
```
