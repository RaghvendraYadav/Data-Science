```python
# Country DataSet
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df_country
```

```python
df['Country Code']
```

```python
# Combine Country Code  from both DatFrame
# merging left on Country Code 
# here Country Column is added
final_df=pd.merge(df,df_country,on='Country Code',how='left')
final_df
```

```python
# Check Data type
df.dtypes
```

```python
# total Records for each country 
final_df['Country'].value_counts()
```

```python
# Getting all the Country Names 
country_names=final_df['Country'].value_counts().index
country_names
```

```python
country_values=final_df['Country'].value_counts().values
```

```python
## Pie Chart
plt.pie(country_values,labels=country_names)
```

```python
## Pie Chart for top three Country tht use Zomato 
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename({0:'Rating Count'},axis=1)
ratings
```

```python
ratings
```

# Conclusions
1 when rating is 4.5 to 4.9 then It is Excellent 
2 when rating is 4.0 to 4.4 then It is Excellent 
3 when rating is 3.5 to 3.9 then It is Good
4 when rating is 3.0 to 3.4 then It is Average
5 when rationg is from 1.0 to 2.4 the it is Poor
6 when rating is below 1.0 then It is Not Rated 

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)

sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
```
