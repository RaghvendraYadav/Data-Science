```python
import pandas as pd
import mysql.connector as conn
import csv

con=conn.connect(host='localhost',user='root',password='0000',database='Task')
cursor=con.cursor()
query="Create Table Attribute_DataSet(Dress_ID int(30),Style varchar(10),Price varchar(10),Rating float(3,1),Size char(5),Season varchar(20),NeckLine varchar(20),SleeveLength varchar(10),waiseline varchar(20),Material varchar(20),FabricType varchar(20),Decoration varchar(20),Pattern_Type varchar(20),Recommendation int(10));"

cursor.execute()
```

```python
data=open('Attribute DataSet1.csv','r')
next(data)
data_csv=csv.reader(data,delimiter=',')
print(data_csv)
for i in enumerate(data_csv):
    for j in i[i]:
        cursor.execute('insert into Task.Attribute_DataSet values({data})'.format(data=(j)))
print('All Data Inserted Successfully')
con.commit()
```
